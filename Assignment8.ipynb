**substr() function**

```r
dbGetQuery(gwasdb, 'SELECT substr(date, 1, 4) AS year FROM gwascat GROUP BY year')
```

add in a call to `avg()` to find the average -log10 p-values per year (using <code>round()</code> to simplify the output) and `count()` to get the number of cases in each group

```r
dbGetQuery(gwasdb, 'SELECT substr(date, 1, 4) AS year,
 round(avg(pvalue_mlog), 4) AS mean_log_pvalue,
 count(pvalue_mlog) AS n
 FROM gwascat GROUP BY year')
```

filtering with `WHERE` applies to rows before grouping  
to filter groups themselves on some condition, you need to use the `HAVING` clause

```r
dbGetQuery(gwasdb, 'SELECT substr(date, 1, 4) AS year,
 round(avg(pvalue_mlog), 4) AS mean_log_pvalue,
 count(pvalue_mlog) AS n
 FROM gwascat GROUP BY year
 HAVING count(pvalue_mlog) > 10')
```

*The type of join (INNER JOIN). The table on the left side of the INNER JOIN statement is known as the left table and the table on the right side is known as the right table.

*The join predicate, which follows ON (in this case, it’s assocs.study_id = studies.id).

*The new notation table.column we use to specify a column from a specific table. This is necessary because there may be duplicate column names in either table, so to select a particular column we need a way to identify which table it’s in.

***Inner joins only select records where the join predicate is true**

```r
dbGetQuery(joinsdb, 'SELECT * FROM assocs INNER JOIN studies ON assocs.study_id = studies.id')
```

```r
dbGetQuery(joinsdb, 'SELECT studies.id, assocs.id, trait, year FROM assocs INNER JOIN studies ON assocs.study_id = studies.id')
```

```r
dbGetQuery(joinsdb, 'SELECT count(*) FROM assocs INNER JOIN studies ON assocs.study_id = studies.id')
```

```r
dbGetQuery(joinsdb, 'SELECT count(*) FROM assocs')
```

```r
dbGetQuery(joinsdb, 'SELECT * FROM studies WHERE id NOT IN (SELECT study_id FROM assocs)')
```

These types of joins will leave some column values NULL (because there’s no corresponding row in the other table) but pair up corresponding records where appropriate. These types of joins are known as outer joins.  
we need all records from the left table, but want to join on data from the right table whenever it’s possible to do so.

```r
dbGetQuery(joinsdb, 'SELECT * FROM assocs LEFT OUTER JOIN studies ON assocs.study_id = studies.id')
```
